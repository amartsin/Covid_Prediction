```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.layers import *
from keras.models import *
from keras.preprocessing import image
```

```python
#CNN MODEL TO PREDICT using Keras
#WE BUILD A SEQUENTIAL MODEL

model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3),activation ='relu',input_shape= (224,224,3)))
# adding 2 layers increase non linearity , so increases hypothesis space thus I am not using a single convolution layer
model.add(Conv2D(64,(3,3),activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25)) #to reduce overfitting
model.add(Conv2D(128,(3,3),activation='relu',))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))


model.add(Flatten())
model.add(Dense(64,activation ='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation = 'sigmoid'))#not softmax because binary classification



#compiling 
model.compile(loss=keras.losses.binary_crossentropy,optimizer = 'adam',metrics='accuracy')
```

```python
#Train
train_datagen= image.ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    #creating horizontal flip which will help in training the model for better accuracy
    horizontal_flip = True
)
test_Dataset=image.ImageDataGenerator(rescale =1./255)
```

```python

def Residual_Unit(input, input_channels, dilation, d):
  for i in range(d):
    x1 = Conv2D(input_channels*2, kernel_size = (1,1), strides = (1,1), padding='same', dilation_rate= (1,1), activation='relu')(input)
    x1 = BatchNormalization()(x1)
  
    a = []

    for i in range(1, dilation+1):
      temp = DepthwiseConv2D( kernel_size=(3,3), dilation_rate = (i,i), padding = 'same', activation= 'relu')(x1)
      temp = BatchNormalization()(temp)
      a.append(temp)

    x = Concatenate(axis= -1)(a)
    x = Conv2D(input_channels, kernel_size = (1,1), strides = (1,1), padding='same', dilation_rate= (1,1), activation='relu')(x)
    x = BatchNormalization()(x)

    x = Add()([x, input])

    input = x
  
  return x
```

```python
# Shifter Unit:

def Shifter_Unit(input, nput_channels, dilation):
    x1 = Conv2D(input_channels*4, kernel_size = (1,1), strides = (1,1), padding='same', dilation_rate= (1,1), activation='relu')(input)
    x1 = BatchNormalization()(x1)

    a = []

    for i in range(1, dilation+1):
      temp = DepthwiseConv2D( kernel_size=(3,3), dilation_rate = (i,i), padding = 'same', activation= 'relu')(x1)
      temp = MaxPool2D(pool_size=(2,2))(temp)
      temp = BatchNormalization()(temp)
      a.append(temp)

    x = Concatenate(axis= -1)(a)

    x = Conv2D(input_channels*2, kernel_size = (1,1), strides = (1,1), padding='same', dilation_rate= (1,1), activation='relu')(x)
    x = BatchNormalization()(x)
```

```python
def Network128(input_shape, nb_class, depth):
  xin = Input(shape= input_shape)

  x = Conv2D(16, kernel_size = (5,5), strides= (1,1), padding = 'same', activation='relu')(xin)
  x = BatchNormalization()(x)

  x = Conv2D(32, kernel_size = (3,3), strides= (2,2), padding = 'same', activation='relu')(x)
  x = BatchNormalization()(x)
  
##Max Dilation rate will be vary in the range (1,5). 

# Max Dilation rate is 5 for tensor (64x64x32)
  x = Residual_Unit(input=x, input_channels=32, max_dilation=5, dir=depth)
  x = Shifter_Unit(input=x, input_channels=32, max_dilation=5)


# Max Dilation rate is 4 for (32x32x64)
  x = Residual_Unit(input=x, input_channels=64, max_dilation=4, d=depth)
  x = Shifter_Unit(input=x, input_channels=64, max_dilation=4)

# Max Dilation rate is 3 for (16x16x128)
  x = Residual_Unit(input=x, input_channels=128, max_dilation=3, d=depth)
  x = Shifter_Unit(input=x, input_channels=128, max_dilation=3)

# Max Dilation rate is 2 for (8x8x256)
  x = Residual_Unit(input=x, input_channels=256, max_dilation=2, d=depth)

  x = GlobalAveragePooling2D()(x)

  x = Dense(64, activation='relu')(x)
  x = Dense(nb_class, activation= 'softmax')(x)

  model = Model(xin, x)

  model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

  return model
```
